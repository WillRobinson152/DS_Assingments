There are 250,894 detached single-family residences in Jacksonville, excluding mobile homes. The median home valuation is $164,661. The median home is 1,640 square feet and 39 years old. 

The most common home feature is a fireplace, which is in nearly 116,000 homes. A pool is the second most popular feature and is present in almost 30,000 homes. Third is a screened porch (27,000 homes), then a covered patio (19,000 homes) and a wooden deck (19,000 homes).

### Construction Trends

Jacksonville's home construction began in the downtown area, along the St. Johns River and along the beaches. Construction has pressed further and further from the river and ocean, with significant recent growth near the airport on the north, around Fleming Island and near the St. Johns County line to the south and along the Intracoastal Waterway to the east.

<div>

</div>

Homes were being constructed larger and larger for a mostly continuous streak for much of the last 70 years. The median home size in 1950 was 1,203 square feet, compared to 1,950 in 2020. However, size has been decreasing since 2016. The median number of bedrooms has remained steady at four for most of the 2000s, while the median number of bathrooms has fluctuated between two and three.

The decrease in size and number of bathrooms in the last couple years may be attributable to home builders like Lennar targeting first-time home buyers and lower income buyers with more affordable homes.

<div>

</div>

Jacksonville saw a housing construction boom in the early 2000s leading up to the real estate crash, followed by a steep crash and gradual rebound. Construction peaked at more than 5,500 in 2005, then bottomed out at 871 in 2011. Almost 3,700 homes were built last year.

### Sales Trends

Home sales were on an upswing coming into 2020. There were 15,642 homes sold last year, generating $3.95 billion - the most sales and the most sale revenue since 2006.

As of July, however, 2020 was slightly behind 2019's pace. This data includes only qualified arms-length transactions of individual properties, a certification process that results in a lag in data.

<div>

</div>

The year started on pace to exceed last year's sales totals. However, sales dropped behind the pace of 2019 in March, then plunged with the onset of the first wave of Covid-19 infections in the U.S. and the economic lockdown. Sales continued to lag in April and May but then bounced back in June. July's sales figures exceeded those netted in 2018 and 2019.

The count of sales and the median price per square foot increased in most Jacksonville zip codes from 2018 to 2019. Sales are highest in North Jacksonville, near the airport, while price per square foot is highest near the beaches and St. Johns County to the south. 

Price per square foot and sales are both low in the downtown area, with the exception of Avondale to the east. This is unsurprising, given that the downtown area has a higher concentration of multifamily housing than single-family housing and has an older inventory of houses.

However, zip codes surrounding the downtown area and continuing south along the St. Johns River were the only zip codes that saw an increase in sales through July on an annual basis. Conversely, the median price per square foot in almost every zip code increased over the same period.
