### Sales Trends

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('DuvalHouseSales.csv')
df.shape
```

```python
reduced = df[['cap_base_yr', 'building_value', 'just_value', 'lot_sf', 'actual_yr_built', 'effec_yr_built',
             'value', 'heated_sf', 'stories', 'bedrooms', 'baths', 'rooms', 'max_units_per_acre', 
             'land_val', 'price', 'fireplace', 'pool', 'deck', 'dock', 'shed', 'sun_rm']]
```

Coming into 2020, Jacksonville's housing market was on an upswing. From 2018 to 2019, sales increased 6.9% to 17,352 and revenue increased 10.3% to $4.3 billion.

<div>

</div>

This year started hot for home sales in Jacksonville. There were hundreds more sales in January and February than in each of the two years prior. March was also the highest in the last three years, despite the onset of Covid-prompted lockdowns at the end of the month. 

However, as lockdowns began in response to the spread of Covid-19 in the U.S., sales fell dramatically. Sales hit bottom in May at 1,097. This likely indicated both a pullback in demand - as layoffs, furloughs and general uncertainty prompted many to delay making large purchases - and constraints on the ability to process sales - as lockdowns impeded the ability to get properties appraised, documents notarized, city fees cleared and the like.

Sales recovered in June and July, but they remained lower than 2019 levels. It is not yet clear if this represents an actual decrease in sales or if sales will stay high further into the fall and winter than is customary, especially since interest rates remain very low.

<div>

</div>

Although sales have lagged since May, sales prices have remained high. The median price per square foot in July was about $7 higher than July 2018 and about $2.30 higher than July 2019. 

This could reflect low interest rates, which enable higher borrowing, as well as a lack of inventory in Jacksonville. Inventory was already low coming into the year, and with the onset of the pandemic, the rate of new listings has declined.

<div>

</div>

The count of sales and the median price per square foot increased in most Jacksonville zip codes from 2018 to 2019. Sales are highest in North Jacksonville, near the airport, while price per square foot is highest near the beaches and St. Johns County to the south. 

Price per square foot and sales are both low in the downtown area, with the exception of Avondale to the east. This is unsurprising, given that the downtown area has a higher concentration of multifamily housing than single-family housing and has an older inventory of houses.

However, zip codes surrounding the downtown area and continuing south along the St. Johns River were the only zip codes that saw an increase in sales through July on an annual basis. Conversely, the median price per square foot in almost every zip code increased over the same period.

<div>

</div>

Homes sold in 2020 often sold for more than their appraised value, as calculated by the county tax assessor. In the graphs below, the orange line depicts a perfect correlation between sales price per square foot and value per square foot.

Trend lines suggest houses are generally selling higher than this line in 2020. For example a house valued at $100 per square foot would sell for $126.08, based on the trend of all home sales in 2020. The relationship is even steeper for homes with certain conditions:

- Homes with river access valued at $100 per square foot would sell for $141.58.
- Homes with pools valued at $100 per square foot would sell for $130.28.
- Homes on golf courses valued at $100 per square foot would sell for $127.65.

<div>

</div>

Homes near water tend to fetch higher prices than those without, but the most valuable body of water to be near is the ocean.

<div>

</div>

Of course, several factors can decrease the price of a house. Intuitively, homes with a high ratio of bedrooms to bathrooms sell for less than if the ratio were closer to 1:1. Similarly, older homes sell for less than newer homes.

<div>

</div>

Building features positively correlated with price include lot size, building size, stories, baths, bedrooms, fireplaces, pools, decks, docks, and sun rooms, among others.
