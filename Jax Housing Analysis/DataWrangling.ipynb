```python
import pandas as pd
```

```python
df = pd.read_csv('JaxHouseSales.csv')
df.fillna(0, inplace=True)
df.head()
```

```python
df.shape
```

```python
empty_cols = []
for i in range(len(df.columns)):
    print(df.columns[i])
    print(df.dtypes[i])
    if df.dtypes[i] != 'object':
        nlls = sum(df[df.columns[i]]==0)
        nll_percent = round(nlls/len(df),2)
        print('Null:', nlls, ' - ', str(nll_percent)+'%')
        if nll_percent >= 0.95:
            empty_cols.append(df.columns[i])
    print()
```

```python
print(empty_cols)
```

```python
df = df.drop(columns=empty_cols)
df.shape
```

```python
for i in range(len(df.columns)):
    print(df.columns[i])
    print(df.dtypes[i])
    print()
```

```python
df = df.astype(
    {'property_use':'object',
     'neighborhood':'object',
     'yr_built':'int64'}
)
```

```python
df.to_csv('JaxHousingData.csv')
```
